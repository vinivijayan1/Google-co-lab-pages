Import Data Set


```python
import pandas as pd
data = pd.read_csv("/content/Assessment_02.csv")
data.head()
```

CHECK INTEGRITY OF DATA

```python

data.info()
```

```python
data.head(20)
```

CALCULATE MEAN VAKUES FOR EACH ATTRIBUTE

```python
mean_MinTemp = round(data['MinTemp'].mean(),2)
mean_MaxTemp = round(data['MaxTemp'].mean(),2)
mean_Rainfall = round(data['Rainfall'].mean(),2)
mean_WindSpeed9am = round(data['WindSpeed9am'].mean(),2)
mean_WindSpeed3pm = round(data['WindSpeed3pm'].mean(),2)
mean_Humidity9am = round(data['Humidity9am'].mean(),2)
mean_Humidity3pm = round(data['Humidity3pm'].mean(),2)
mean_Temp9am = round(data['Temp9am'].mean(),2)
mean_Temp3pm = round(data['Temp3pm'].mean(),2)
```

REPLACE NULL VALUES WITH MEAN

```python
data['MinTemp'] = data['MinTemp'].fillna(mean_MinTemp)
data['MaxTemp'] = data['MaxTemp'].fillna(mean_MaxTemp)
data['Rainfall'] = data['Rainfall'].fillna(mean_Rainfall)
data['WindSpeed9am'] = data['WindSpeed9am'].fillna(mean_WindSpeed9am)
data['WindSpeed3pm'] = data['WindSpeed3pm'].fillna(mean_WindSpeed3pm)
data['Humidity9am'] = data['Humidity9am'].fillna(mean_Humidity9am)
data['Humidity3pm'] = data['Humidity3pm'].fillna(mean_Humidity3pm)
data['Temp9am'] = data['Temp9am'].fillna(mean_Temp9am)
data['Temp3pm'] = data['Temp3pm'].fillna(mean_Temp3pm)
```

```python
data.head(20)
```

RECHECK DATA INTEGRITY

```python
data.info()
```

ENCODE TEXT VALUES WITH NUMERICAL VALUES
NaN replaced with NULL, and then mapped to 0.5 (uncertain)


```python
data['RainToday'] = data['RainToday'].fillna("NULL")
data['RainToday'] = data['RainToday'].map({"Yes":1,"No":0, "NULL":0.5})
data.head(20)
```

CHECK DATA IS ALL COMPLETE

```python
data.info()
```

ENCODE CATEGORICAL DATA IN DEPENDENT VARIABLE 

```python
data['RainTomorrow'] = data['RainTomorrow'].map({"Yes":1,"No":0})
data.head(20)
```

CREATE NEW DATA FRAME- DISREGARD INDEX AND LOCATION


```python
X = data.iloc[:,2:13]
y = data.iloc[:,13]
```

```python
X
```

```python
y
```

IMPLEMENT TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.80)
```

```python
len(X_train)
```

```python
len(X_test)
```

SCALE THE INDEPENDENT VARIABLES

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

CREATE THE MODEL


```python
import keras as tf

model = tf.models.Sequential()

model.add(tf.layers.Dense(8,input_dim=11,activation='tanh'))    ## input layer of 11 and one first hidden layer of 8 using tanh function
model.add(tf.layers.Dense(16,activation='tanh'))                ## second hidden layer of 16 using tanh function
model.add(tf.layers.Dense(1,activation='sigmoid'))              ## output layer of 1 for binary classifier using sigmoid function

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy']) # stochastic gradient binary crossentropy learning curve for Output layer of 1
```

```python
model.fit(X_train,y_train,epochs=100)
```

Evaluate Model

```python
model.evaluate(X_test,y_test)
```

```python
ypred = model.predict(X_test)

ypred
```

Create ypred values for Confusion Matrix

```python
ypred = (ypred > 0.5) # 50% threshold for Confusion Matrix
ypred                 # shows ypred in correct data type for confusion matrix
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,ypred)
cm
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(y_test,ypred)
a
```

Predicting a single Value- Row 100 from the dataset


```python
import numpy as np
#inputs = [15.5, 32, 0, 2, 4, 51, 25, 20.1, 30.8, 0, 1.2 ]
new_prediction = model.predict(sc.transform(np.array([[15.5, 32, 0, 2, 4, 51, 25, 20.1, 30.8, 0, 1.2]])))
new_prediction = (new_prediction > 0.5)
print(new_prediction)
```
