```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
import keras as tf
import csv

def row_count(file):
    with open(file) as f1:
        # data = [row for row in csv.reader(f1)]
        row_count = sum(1 for row in csv.reader(open(file)))
        # print(row_count)
        return (row_count)

#data=pd.read_csv('NN_input.csv')
data=pd.read_csv('/output1.csv')
print(data.info())

with open('/output1.csv') as f:
    reader = csv.reader(f)
    data1 = [tuple(row) for row in reader]
    length = len(data1[0])
    print('length-->', length)



x=data.iloc[:,0:length-1]
y=data.iloc[:,length-1]

###split the data into train and testing sample
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.80,random_state=2)

print('xtrain', len(xtrain))
print('xtest', len(xtest))
print('ytrain', len(xtrain))
print('ytest', len(xtest))
print('xtest-->', xtest)
print('ytest-->', ytest)

np.unique(ytrain,return_counts=True)
model=tf.models.Sequential()#blank model

#add the layers
model.add(tf.layers.Dense(32,input_dim=length-1,activation='relu' ))
model.add(tf.layers.Dense(32,activation='relu'))
model.add(tf.layers.Dense(32,activation='relu'))

model.add(tf.layers.Dense(2,activation='sigmoid'))
#model.add(tf.layers.Dense(2,activation='softmax'))
#model.add(tf.layers.Dense(1,activation='tanh'))

#model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
#model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
#model.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
#model.compile(loss='mean_squared_logarithmic_error',optimizer='adam',metrics=['accuracy'])
#model.compile(loss='hinge',optimizer='adam',metrics=['accuracy'])
#model.compile(loss='squared_hinge',optimizer='adam',metrics=['accuracy'])
#model.compile(loss='kullback_leibler_divergence',optimizer='adam',metrics=['accuracy'])

model.fit(xtrain,ytrain,epochs=500)

### Testing / Predictions
ypred = model.predict(xtest)
#ypred=ypred.round()


new_ypred = []
for i in range(len(ypred)):
  new_ypred.append(ypred[i].argmax())
new_ypred = np.array(new_ypred)
print(new_ypred)

print('y.test.shape-->', ytest.shape)
print('y.pred.shape-->', new_ypred.shape)

model.evaluate(xtest,ytest)

cm=confusion_matrix(ytest, new_ypred)
#cm1=cm.transpose()
print(cm)

a = accuracy_score(ytest, new_ypred)
print(a)


input=[4.979549885,-4.923600197,-4.979549885,-4.811699867,-4.755750179,-4.587900162,-4.475999832,-3.692699909,-0.671400011,3.860549927,9.56744957,15.05055046,19.13489914,21.82049942,22.99544907,23.21924973,22.43594933,20.81340027,18.5753994,16.11359978,13.76369953,11.74950027,10.18290043,8.896050453,7.833000183,6.602099895,5.98664999,6.042600155,5.98664999,5.70690012,6.098549843,6.322350025,6.546150208,6.714000225,6.602099895,6.769949913,7.385399818,8.112750053,8.840100288,10.2947998,12.25304985,14.8267498,17.62425041,20.58959961,23.16329956,25.62509918,27.69524956,29.31780052,30.54870033,31.27605057,31.05224991,30.32489967,29.26185036,27.80714989,25.79294968,23.16329956,19.86224937,16.11359978,11.74950027,7.105649948,2.965349913,-0.335700005,-2.741549969,-3.916500092,-4.867650032,-5.59499979,-6.37830019,-7.161600113,-7.944900036,-10.68645,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973,-14.15534973]
print(model.predict([input]).argmax())

input1=[5.83144286556,7.35774819705,-3.43509951891,7.5446033544,-5.21031601778,9.20638125991,36.8735795166,23.445399888,32.675719752,32.6285748844,23.1071439699,34.802948013,28.6616698688,-1.87397637967,14.4866931404,3.48551121642,28.5235661034,39.8504127035,-7.04713364425,39.2171133237,12.9297833188,22.4245152352,8.46015568712,12.1584494771,-1.19271790357,20.5183705899,31.3973324285,-5.78778239278,39.4685486907,28.5447678919,17.5141846204,-7.45715616554,30.426020043,32.7846995559,16.8215368775,24.0122443529,34.0275173534,29.7690607552,18.5067182385,-9.20423302379,-5.15247405473,-2.05801143401,-7.06129229973,10.6324626365,-5.64743677874,2.88825947353,-2.75261016916,-2.42244914132,36.9155448393,1.35770980803,9.67689111158,-0.222655244345,13.782266275,28.8562290584,7.86654821983,1.37930122115,15.0546098702,-2.61220766387,-6.56931730572,-9.92137820047,21.6689981914,11.1529522356,25.0051540226,-5.95352177882,39.624483421,4.45175182225,39.925587925,14.6655175751,3.89210741736,15.0222197002,11.0642325286,10.0440071189,30.11541776,19.9985455093,9.38568750874,-8.10507441801,11.6100298997,26.2791415282,26.9522348405,35.4543724852,6.13624254757,12.2800491972,27.2853537246,7.94037021378,-6.98147587759,14.7150603683,1.05475727696,-8.95918467619,8.73581833298,33.4680550849,35.4462923002,-5.83112068516,29.6488601259,-0.996291700025,11.3479646325,33.9743513432,19.40114276,25.2698271293,31.6344163672,35.8371642115,20.273917013,-0.7047325221,10.828920252,10.2741399931,26.4213275569,29.9135119017,31.0382980373,29.3955087099,-0.977117929198,-4.92139297662,-2.12649938986,-3.77270961596,37.0111287987,-1.47542436504,38.0639771396,2.6764890904,10.6859691009,4.17029538318,-3.56630204277]
print(model.predict([input1]).argmax())

model.save('NN_30cst')
```
