```
import pandas as pd
```

```
data = pd.read_csv("/content/breast_cancer.csv")
data.head()
```

```
data.info()
```

```
data['diagnosis'].value_counts()
```

```
data['new_diagnosis'] = data['diagnosis'].map({"M":1,"B":0})
data.head()
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['new_diagnosis'] = le.fit_transform(data['diagnosis'])
data.head()
```

```
data.head(1)
```

```
data.info()
```

```
### divide the data into features and targets
x = data.iloc[:,2:32]
y = data.iloc[:,33]
```

```
### Split the data into train and testing samples
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.80,random_state=1)
```

```
len(xtrain)
```

```
len(xtest)
```

```
data['diagnosis'].value_counts()
```

```
import numpy as np
np.unique(ytrain,return_counts=True)
```

```
import keras as tf
```

```
model = tf.models.Sequential()  ## blank model
## add the layers
model.add(tf.layers.Dense(16,input_dim=30,activation='relu'))  ## input and one hidden layer
model.add(tf.layers.Dense(32,activation='relu'))  ## second hidden layer
model.add(tf.layers.Dense(2,activation='softmax'))
## Compile the model
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
### binary_crossentropy | sparse_categorical_crossentropy
```

```
### Training the model
model.fit(xtrain,ytrain,epochs=500)
```

```
### Testing / Predictions
ypred = model.predict(xtest)
ypred
```

```
ypred[1].argmax()
```

```
new_ypred = []
for i in range(len(ypred)):
  new_ypred.append(ypred[i].argmax())
```

```
ytest.shape
```

```
import numpy as np
new_ypred = np.array(new_ypred)
new_ypred
```

```
ypred
```

```
model.evaluate(xtest,ytest)
```

```
### Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,new_ypred)
cm
```

```
### Confusion Matrix
from sklearn.metrics import accuracy_score
a = accuracy_score(ytest,new_ypred)
a
```
