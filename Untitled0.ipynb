```
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import keras as tf

data=pd.read_csv('/breast_cancer.csv')
data.head()
le=LabelEncoder()
data['new-diagnosis']=le.fit_transform(data['diagnosis'])
x=data.iloc[:,2:32]
#x.info()
y=data.iloc[:,33]
###split the data into train and testing sample
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.80,random_state=1)

np.unique(ytrain,return_counts=True)
model=tf.models.Sequential()#blank model
#add the layers
model.add(tf.layers.Dense(16,input_dim=30,activation='relu'))
model.add(tf.layers.Dense(32,activation='relu'))
model.add(tf.layers.Dense(1,activation='softmax'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(xtrain,ytrain,epochs=500)

### Testing / Predictions
ypred = model.predict(xtest)
ypred = ypred.round()
#ypred
#ytest

```
